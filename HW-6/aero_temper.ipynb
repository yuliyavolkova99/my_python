```python
with open('temper.txt', 'r') as t:
    lst = t.readlines()
    out = []
    for i in lst:
        out.append(float(i.strip()))
print('Максимальная температура:', max(out))
print('Минимальная температура:', min(out))
print('Количество уникальных температур:', len(list(set(out))))
print('Средняя температура:', sum(out)/len(out))

import matplotlib.pyplot as plt
y_coords = out
plt.plot(y_coords)
plt.show()
print('График изменения температур в градусах по Фаренгейту одного из аэропортов мира в период с 1948 по 2016')
```
